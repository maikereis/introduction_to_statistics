```python
import numpy as np
import pandas as pd
from scipy.linalg import cholesky
from scipy.stats import multivariate_normal as mvn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
sns.set_palette("Blues_r")
sns.set(font_scale=1.20, rc={'figure.figsize':(12,6)})
sns.set_context(rc={"axes.facecolor": ".9", "axes.titlesize":16,"axes.labelsize":15})
sns.despine()
```

```python
# Correlation matrix
cov= np.array([[   1, 0.6],
                [0.6, 1]])
scores = mvn.rvs(mean = [50.,50], cov=cov, size = 100)
data = pd.DataFrame(scores)
```

## The regression line

If the scatterplot shows a linear association, then this relationship can be summarized by a line.

```python

fig, axs = plt.subplots(1,2)

g1 = sns.scatterplot(x=data[0], y=data[1], ax=axs[0])
g2 = sns.regplot(x=data[0], y=data[1], ax=axs[1], line_kws={"color": "red"})

g1.set_title(f"$r$={data[0].corr(data[1]):0.2f}")

g1.set(xlabel=None, ylabel=None)
g2.set(xlabel=None, ylabel=None)

    
plt.tight_layout()
```

To find this line for $n$ pairs of data ($x_1, y_1$),...., ($x_n, y_n$), recall that the equation of a line produces the y-value $\hat{y}_i = a + bx_i$. The idea is to choose the line that minimizes the sum of the squared distances between the observed $y_i$ and the $\hat{y}_i$. In other words, for $n$ pairs of data ($x_1, y_1$),...., ($x_n, y_n$), find $a$ and $b$ that minimize

$$\sum_{i=1}^{n}{(y_i - \hat{y}_i)^2}= \sum_{i=1}^{n}{(y_i - (a + bx_i))^2} $$

This is the **method os least squares**. It turn out that $b=r\frac{s_y}{s_x}$ and $a = \tilde{y}-b\tilde{x}$. This line $\hat{y}=a+bx$ is called the **regression line**.

There is another interpretation of the regression line:
it computes the average value of $y$ when the first coordinate is near $x$.

Remenber that often times an average is the 'best' predictor. This shows how the regression line incoporates the information given by $x$ to prouce a good preictor of $y$.
