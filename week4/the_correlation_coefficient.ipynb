## The Correlation Coefficient

The scatterplot visualizes the relationship between two quantitative variables. It may have a direction (sloping up or down), form (a scatter that clusters around a line is called linear) and strength (how closely do the point follow the form?).

```python

ax = sns.regplot(x=heights['Father'], y=heights['Son'], line_kws={"color": "red"})
_ = ax.set_title("Father's and Son's heights relationship")
```

If the form is linear, then a good measure of strengh is the **correlation coefficient r**:<br>
Our data are($x_i, y_i$), $i$= 1,...,$n$.

$$r = \frac{1}{n}\sum_{i=1}^{n}{\frac{x_i - \tilde{x}}{s_x}} \times {\frac{y_i - \tilde{y}}{s_y}}$$ 

(divide by $n-1$ instead of $n$ if this is done for the standard deviations $s_x, s_y$).

*In the case where the scatter clusters around a line*, it's very useful to summarize that clustering with correlation coefficient **r**. 

The correlation coefficient measures the **strenght** and the **direction** of relatioship.
