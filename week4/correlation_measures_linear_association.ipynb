```python
import numpy as np
import pandas as pd
from scipy.linalg import cholesky
from scipy.stats import multivariate_normal as mvn

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
sns.set_palette("Blues_r")
sns.set(font_scale=1.20, rc={'figure.figsize':(8,6)})
sns.set_context(rc={"axes.facecolor": ".9", "axes.titlesize":16,"axes.labelsize":15})
sns.despine()
```

```python
# Correlation matrix
cov= np.array([[   1, -0.9, -0.6,   0, 0.2],
               [-0.9,    1,  0.5, 0.5,    0],
               [-0.6,  0.5,    1,   0,    0],
               [   0,  0.5,    0,   1,    0],
               [0.2,    0,    0,   0,    1]])

scores = mvn.rvs(mean = [50.,50.,50.,50., 50.], cov=cov, size = 500)

data = pd.DataFrame(scores)
```

## Correlation Measures Linear Association

A numerical summary of these pais of data is given by: $\tilde{x}, s_x, \tilde{y}, s_y, r$.

As a convetion the variable on the horizontal axis is called **explanatory variable** or **predictor**, the one on the vertical axis is called **response variable**.

$r$ is aways between -1 and 1. The sign of $r$ gives the direction of the association and its absolute value gives the strengh:

Since both $x$ and $y$ were standardized when computing $r$, $r$ has no units and is not affected by changing the center or the scale of either variable.


```python

fig, axs = plt.subplots(2,2)

for i in range(0, 5):
    ax = plt.subplot(2, 3, i+1)
    g = sns.regplot(x=data[0], y=data[i], ax=ax, line_kws={"color": "red"})
    g.set_title(f"$r$={data[0].corr(data[i]):0.2f}")
    g.set(xlabel=None, ylabel=None)
    
plt.tight_layout()
```

Keep in mind that $r$ is only useful for measuring *linear* association
