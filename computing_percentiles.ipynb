## Computing Percentiles for normal data

What is the 30th percentile of the father's heights?

$z = ?$

```python

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.arange(mu - 4*sigma, mu + 4*sigma, 0.01)
ax = sns.lineplot(x=x, y=stats.norm.pdf(x, mu, sigma))
kde_x, kde_y = ax.lines[0].get_data()
a_1 = (-1.5)

ticks_loc = ax.get_xticks().tolist()
ax.set_xticks(ticks_loc)
ax.set_xticklabels([f'{x:,.0f}σ' for x in ticks_loc])

ax.annotate("What is the value\n where 30% of the data\n fall below?", xy=(-1.5, 0.00),  xycoords='data',
            xytext=(0.5, 0.2), textcoords='axes fraction',
            arrowprops=dict(facecolor='gray', shrink=0.1),
            horizontalalignment='center', verticalalignment='center',
            )

ax.fill_between(kde_x, kde_y, 
where=(kde_x < a_1) , 
interpolate=True, color='#85756d', alpha=0.50)
```
