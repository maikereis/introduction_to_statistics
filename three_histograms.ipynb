## Three Histograms

The chance process of tossing a coin 100 times comes with three different histograms:

1. The probability histogram for producing the data:

```python

ax = sns.barplot(x=['Heads','Tails'], y=[0.5,0.5])
_ = ax.set_title("The probability histogram")
```

2. The histogram of the 100 observed tosse. This is an empirical histogram of real data:

```python

ax = sns.barplot(x=['Heads','Tails'], y=[0.47,0.53])
_ = ax.set_title("The probability histogram")
```

3. The probability histogram of the statistic $S_{100}$ = 'number of tails', which shows the sampling distribution of $S_{100}$:

```python

fig, ax = plt.subplots(1, 1)

n, p = 100, 0.5

x = np.arange(0, n+1)
pmf_x = binom.pmf(x, n, p)

ax.vlines(x, 0, binom.pmf(x, n, p), lw=4, alpha=0.7)
ax.set_xlabel(r'Number of successes ($X$)')
ax.set_ylabel(r'Probability')
g = ax.plot(x, pmf_x, ms=8)
```
