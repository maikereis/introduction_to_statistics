## Random Variables

The outcomes of the 10 experiments are due to chance, so the number of success of random: one set of 10 experiments might result of 4 success, another set might result in 7 successes.

X = 'number of successes' is called a **random variable**.

$P$(X=2)$=30.2\%$

$\space X$ has the **binomial distribution**.

We can visualize the probabilities of the various outcomes of $X$ with a probability histogram:

```python

fig, ax = plt.subplots(1, 1)

n, p = 10, 0.2

x = np.arange(0, n+1)
pmf_x = binom.pmf(x, n, p)

ax.vlines(x, 0, binom.pmf(x, n, p), lw=4, alpha=0.7)
ax.set_xlabel(r'Number of successes ($X$)')
ax.set_ylabel(r'Probability')
g = ax.plot(x, pmf_x, 'go', ms=8)
```

A histogram of data gives percentages for observed data. In contrast, a probability histogram is a theoretical construct: it visualizes probabilities rather than data that have been empirically observed.
