## Standardizing Data and the Standard Normal Curve

A normal curve is determined by $\tilde{x}$ and $s$; if the data follows the normal curve, then knowing $\tilde{x}$ and $s$, means knowing the whole histogram.

To compute areas under the normal curve, we first standardize the data by subtracting off $\tilde{x}$ and then dividing by $s$:

$$z= \frac{height - \tilde{x}}{s}$$

$z$ is called the standardized value or $z$-score.

$z$ has no unit (height, $\tilde{x}$ and $s$ all have unit 'inches')

```python
mean, std = hdf.mean(), hdf.std()
```

The $z$-score means the number of standard deviations

```python
z = ( hdf - mean / std )
z.head(5)
```

## The Standard Normal Curve

Standardized data have mean 0 and standard deviation equal to 1 - this is the point of standardizing.

The father's heights follows the normal curve with $\tilde{x} = 68.3$ in and  ${s} = 1.8$ in. Therefore the standardized values follow the **standard normal curve** with $\tilde{x} = 0 $ and $s = 1.8$  

Normal curve or gaussian is given by the formula:

$$f(x)=\frac{1}{\sigma\sqrt(2\pi)}e^{\frac{1}{2}(\frac{x-\mu}{\sigma})^{2}}$$

or, 

$$X\sim{}N(0,1)$$

```python

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.arange(mu - 4*sigma, mu + 4*sigma, 0.01)
ax = sns.lineplot(x=x, y=stats.norm.pdf(x, mu, sigma))

ticks_loc = ax.get_xticks().tolist()
ax.set_xticks(ticks_loc)
ax.set_xticklabels([f'{x:,.0f}σ' for x in ticks_loc])

ax.text(0.8, 0.8, r"$f(x)={\frac{1}{\sqrt{2\pi}}}e^{\frac{1}{2}x^{2}}$", horizontalalignment='center',
     verticalalignment='center', transform=ax.transAxes, fontsize=20, fontfamily='serif', fontstyle='oblique')

ax.axvline(0, color = 'k', linestyle="--")
ax.annotate(f'$\mu$ = 0', xy=(0.1, 0.2))
```
